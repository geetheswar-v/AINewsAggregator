```python
import pandas as pd
```

```python
# Data Preprocessing
# removing extra description which is more than 30 word length
# before training the model
articles = pd.read_csv('cleaned_articles.csv')

articles.head()
```

```python
articles.tail()
```

```python
articles = articles.drop_duplicates(subset=['Title'])
articles.describe()
```

```python
def preprocess_text(text):
    if isinstance(text, str):
        words = text.split(' ')
        if len(words) > 30:
            return ' '.join(words[:31])
        return text
    else:
        print(text)
        return text
```

```python
articles = articles[articles['Description'].notna()]
articles.reset_index(drop=True, inplace=True)
articles.describe()
```

```python
articles['Title'] = articles['Title'].apply(preprocess_text)
articles['Description'] = articles['Description'].apply(preprocess_text)
```
